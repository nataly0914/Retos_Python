# Reto Visualización con Matplotlib - Visualización Basica

En este reto haremos visualizaciones con la librería Matplotlib de python, la cual es muy fácil e intuitiva de usar. Estaremos trabajando principalmente con gráficos de líneas

1- Vamos a hacer un gráfico simple que representa el gasto de alguien en el almuerzo durante la semana pasada. Primero, defina dos listas, `days` y `money_spent`, que contengan los siguientes números enteros:



````python
import matplotlib.pyplot as plt
days = (0,1,2,3,4,5,6)
money_spent = (10,12,12,10,14,22,24)
print(days)
print(money_spent)
````

2- Grafica `days` en el eje `x` y `money_spent` en el eje `y` usando `plt.plot()`. Mostrar el gráfico usando `plt.show()`.

````python
plt.plot(days,money_spent)
plt.ylabel('Money spent')
plt.xlabel('Days')
plt.title('Gráfico de líneas')
plt.grid(True)
plt.show()
````

3- Hemos definido listas llamadas `time`, `revenue` y `costs`. 

```
time = [0, 1, 2, 3, 4]
revenue = [200, 400, 650, 800, 850]
costs = [150, 500, 550, 550, 560]
```

Copielas en la siguiente celda

````python
time = [0, 1, 2, 3, 4]
revenue = [200, 400, 650, 800, 850]
costs = [150, 500, 550, 550, 560]
print(time)
print(revenue)
print(costs)
````

4- Grafica el `revenue` vs. `time`.

````python
plt.plot(revenue,time, color='red')
plt.ylabel('Time')
plt.xlabel('Revenue')
plt.show()
````

5- Grafica `costs` vs `time` en el mismo grafico del `time` vs `revenue`

````python
#línea 1
plt.plot(revenue,time, color='red')

#línea 2
plt.plot(costs,time, color='green')
plt.ylabel('Time')
plt.xlabel('Costs / Revenue')
plt.show()
````

6- Grafique `revenue` vs `time` como una linea purple (`purple`), y dashed (`--`).

````python
plt.plot(revenue,time, color='purple', ls='dashed', lw=2)
plt.ylabel('Time')
plt.xlabel('Revenue')
plt.show()
````

7- Grafique `costs` vs `time` como una linea con color HEX `#82edc9` y marcadores cuadrados (`s`)

````python
plt.plot(costs,time, color='#82edc9', marker='s')
plt.ylabel('Time')
plt.xlabel('Costs')
plt.show()
````

8- Grafique los dos anteriores en un mismo grafico

````python
#línea 1
plt.plot(revenue,time, color='purple', ls='dashed', lw=2)

#línea 2
plt.plot(costs,time, color='#82edc9', marker='s')
plt.ylabel('Time')
plt.xlabel('Costs / Revenue')
plt.show()
````

9- Hemos trazado una línea que representa el gasto de alguien en café durante los últimos 12 años. Corra el codigo para ver el resultado.

```
x = range(12)
y = [3000, 3005, 3010, 2900, 2950, 3050, 3000, 3100, 2980, 2980, 2920, 3010]

```

````python
x = range(12)
y = [3000, 3005, 3010, 2900, 2950, 3050, 3000, 3100, 2980, 2980, 2920, 3010]
plt.plot(x, y, color = 'orange')
plt.ylabel('Gasto en café')
plt.xlabel('Años')
plt.show()
````

10- Modifiquemos los ejes para ampliar un poco más nuestro gráfico de líneas. Utilice `plt.axis()` para modificar los ejes para que el eje `x` vaya de `0` a `12`, y el eje `y` vaya de `2900` a `3100`.

````python
plt.plot(x, y, color = 'orange')
plt.ylabel('Gasto en café')
plt.xlabel('Time')
plt.axis([0, 12, 2900, 3100])
plt.show()
````

11 - Marca el eje `x` como `Time` asignando la leyenda

````python
plt.plot(x, y, color = 'orange')
plt.ylabel('Gasto en café')
plt.xlabel('Time')
plt.axis([0, 12, 2900, 3100])
plt.show()
````

12- Marca el eje `y` como `Dollars spent on coffee` asignando la leyenda correspondiente

````python
plt.plot(x, y, color = 'orange')
plt.ylabel('Dollars spent on coffee')
plt.xlabel('Time')
plt.axis([0, 12, 2900, 3100])
plt.show()
````

13- Añade el titulo `My Last Twelve Years of Coffee Drinking`

````python
plt.plot(x, y, color = 'orange')
plt.ylabel('Dollars spent on coffee')
plt.xlabel('Time')
plt.title('My Last Twelve Years of Coffee Drinking')
plt.axis([0, 12, 2900, 3100])
plt.show()
````
