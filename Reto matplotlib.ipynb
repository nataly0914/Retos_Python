```python
import matplotlib.pyplot as plt
```

```python
days = (0,1,2,3,4,5,6)
money_spent = (10,12,12,10,14,22,24)
```

```python
print(days)
print(money_spent)
```

```python
plt.plot(days,money_spent)
plt.ylabel('Money spent')
plt.xlabel('Days')
plt.title('Gráfico de líneas')
plt.grid(True)
plt.show()
```

```python
time = [0, 1, 2, 3, 4]
revenue = [200, 400, 650, 800, 850]
costs = [150, 500, 550, 550, 560]
```

```python
plt.plot(revenue,time, color='red')
plt.ylabel('Time')
plt.xlabel('Revenue')
plt.show()
```

```python
#línea 1
plt.plot(revenue,time, color='red')

#línea 2
plt.plot(costs,time, color='green')
plt.ylabel('Time')
plt.xlabel('Costs / Revenue')
plt.show()
```

```python
plt.plot(revenue,time, color='purple', ls='dashed', lw=2)
plt.ylabel('Time')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(costs,time, color='#82edc9', marker='s')
plt.ylabel('Time')
plt.xlabel('Costs')
plt.show()
```
