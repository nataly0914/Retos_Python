```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
data1 = pd.read_csv("http://rcs.bu.edu/examples/python/data_analysis/flights.csv")
```

```python
#seleccionar datos que tengan al menos 1 na
data1[data1.isnull().any(axis=1)].head()
```

```python
data2 = data1[['carrier','air_time','origin', 'dest', 'distance']]
print(data2)
data2.dropna()
data2.dropna(how='all')
```

```python
grafica = sns.displot(data=data1['dep_time'], color='red')
grafica.set_axis_labels ('Hola de partida', 'Conteo')
grafica.set_titles('Distribución de las horas de partida')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.boxplot(x=data1['distance'], y=data1['carrier'], linewidth=2.0, orient="h")
plt.show()
```

```python
sns.scatterplot(data=data1, x="distance", y="air_time", color='green', hue="carrier")
```
